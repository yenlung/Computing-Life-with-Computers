#　在 Jupyter (IPython) 中用 R

## 1. 讓 IPython 可以和 R 溝通

先期準備

* 要安裝好 R (當然)
* 要有 ryp2 套件 (我是用 pip install)

###[注意]

很多人都是寫

    %load_ext rmagic

但新版 ryp2 已改成:

    %load_ext rpy2.ipython

```python
%load_ext rpy2.ipython
```

## 2. 直接來一段 R

如果用 %%R, 那整個 cell 都給 R 了!

以下的一小段是從 Code School 的 R 教學來的, 有興趣學 R 的可以多加例用, 很快上手!

[Try R](http://tryr.codeschool.com)

```python
%%R
A <- matrix(1, 15, 15)
A[4,7] <- 0
persp(A, expand=0.5)
```

##3. 只有一小段給 R

如果用 %%R 是整個 cell 給 R。但只有一句話要給 R, 我們可以用 %R。

```python
%R print("Hi from R")
print("Hi from Python")
```

##4. 從 R 要資料

本來 Python 是沒有 A 的。

```python
A
```

向 R 要 A!!

```python
%R -o A
```

```python
A
```

```python
A.dim
```

很容易轉成 numpy 的 array, 形式保留。

```python
from numpy import array
```

```python
A = array(A)
```

```python
A.shape
```

##5. 從 Python 送資料給 R

```python
from numpy import linspace, sin
```

```python
x = linspace(0, 10 , 100)
y = sin(x) + 0.5*x
```

```python
%R -i x
```

```python
%R -i y
```

```python
%R plot(x,y)
```
